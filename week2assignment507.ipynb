## Indentifying and Handling the Missing Values in the Dataset

  - After that used the data.fillna function to fill in the missing values as NaN (Not a Number), which I also used when loading the CSV file to indicate where in the dataset there were missing values that were expressed by "missing"

```python
missing_values = hcdata.isnull()
all_missing_values = missing_values.sum() .sum()
print(all_missing_values)
```

```python
missing_values = hcdata.fillna(np.nan)
missing_values
```

### Removing Duplicate Rows and Columns/ Cleaning Columns

  - I removed duplicate rows by using the .dropna(inplace = True) function
  - I removed the duplicate columns by using the .loc [:, ~hcdata.columns.duplicated()] function. The source I used to help me with this function was https://sparkbyexamples.com/pandas/pandas-remove-duplicate-columns-from-dataframe/
  - To clean the columns names I used hcdata.columns = hcdata.columns.str.strip()
  - Additionally, to clean the columns I used the clean_value function to remove any white space from the columns

```python
hcdata.dropna(inplace = True)
hcdata
```

```python
hcdata = hcdata.loc [:, ~hcdata.columns.duplicated()]
hcdata
```

```python
hcdata.columns = hcdata.columns.str.strip()
hcdata
```

```python
# Function to remove white space and special characters from a value
def clean_column_names(hcdata):
    # Define a helper function to clean column names
    def clean_name(value):
        cleaned_name = re.sub(r'[^a-zA-Z0-9]', '', value)
        return cleaned_name.lower()


    hcdata.columns = [clean_name(col) for col in hcdata.columns]
    return hcdata

data = clean_column_names(hcdata)

print(hcdata)
```
